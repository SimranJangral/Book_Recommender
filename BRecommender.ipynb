##Popularity Based Recommender System.
Our system will recommend TOP 25 books which have highest rating provided more than 400 users have given ratings to that book.




```python
ratings_with_name = ratings.merge(books, on='ISBN')
ratings_with_name.head()
```

```python
num_rating_df= ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'Num_ratings'}, inplace=True)
num_rating_df
```

```python
ratings_with_name.head()
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Num_ratings']>=400].sort_values('avg_rating',ascending=False).head(25)
```

```python
popular_df= popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Num_ratings','avg_rating']]
popular_df
```

```python
popular_df.shape
```

##Colaborative filtering system

```python
x= ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
frequent_readers = x[x].index
#this will give the Users and how many ratings they have done
```

```python
filtered_rating =ratings_with_name[ratings_with_name['User-ID'].isin(frequent_readers)]
```

```python
y= filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books =y[y].index
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
plot =final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
plot.fillna(0,inplace=True)
plot
```

```python
import sklearn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(plot)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(plot.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similar_items:
        item =[]
        temp_df = books[books['Book-Title'] == plot.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('1984')
```
